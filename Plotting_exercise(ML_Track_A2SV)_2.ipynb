```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('imdb_top_1000.csv')

# Convert the 'year' column to numeric type
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')

# Group the DataFrame by year and count the number of movies
movies_per_year = df.groupby('Released_Year')['Series_Title'].count()

# Plot a scatterplot
plt.scatter(movies_per_year.index, movies_per_year.values)

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Number of Series')
plt.title('Number of Series vs Year')

# Display the plot
plt.show()
```
